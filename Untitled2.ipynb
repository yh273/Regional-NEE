```python
# 1. 导入库和加载数据
import xarray as xr
import matplotlib.pyplot as plt

# 加载GPP数据
store = 'gs://leap-persistent-ro/data-library/feedstocks/metaflux_feedstock/meatflux_monthly.zarr'
ds = xr.open_dataset(store, engine='zarr', chunks={})

# 加载LAI数据  
store_lai = 'https://nyu1.osn.mghpcc.org/leap-pangeo-pipeline/MODIS_LAI/MODIS_LAI.zarr'
LAI = xr.open_dataset(store_lai, engine='zarr', chunks={})

# 2. 定义预处理函数
def preprocess_lai_for_alaska(gpp_ds, lai_ds, lat_ak=65.0, lon_ak=-150.0):
    """为阿拉斯加格点预处理LAI数据"""
    
    # 1. 时间聚合
    lai_monthly = lai_ds['lai'].resample(time='1M').mean()
    
    # 2. 空间重采样到GPP网格
    lai_regridded = lai_monthly.interp(
        lat=gpp_ds.latitude,
        lon=gpp_ds.longitude,
        method='linear'
    )
    
    # 3. 时间对齐
    common_time = slice('2002-01-01', '2020-12-31')
    lai_aligned = lai_regridded.sel(time=common_time)
    gpp_aligned = gpp_ds.sel(time=common_time)
    
    # 4. 提取阿拉斯加格点
    ak_lai = lai_aligned.sel(
        latitude=lat_ak, 
        longitude=lon_ak, 
        method='nearest'
    )
    ak_gpp = gpp_aligned.sel(
        latitude=lat_ak, 
        longitude=lon_ak, 
        method='nearest'
    )
    
    return ak_lai, ak_gpp

# 3. 调用函数
lat_ak = 65.0
lon_ak = -150.0
ak_lai, ak_gpp = preprocess_lai_for_alaska(ds, LAI, lat_ak, lon_ak)

# 4. 现在运行检查（这时ak_lai和ak_gpp已经定义了）
print("=== 时间维度检查 ===")
print(f"LAI 时间范围: {ak_lai.time.min().values} 到 {ak_lai.time.max().values}")
print(f"GPP 时间范围: {ak_gpp.time.min().values} 到 {ak_gpp.time.max().values}")
print(f"时间点数量 - LAI: {len(ak_lai.time)}, GPP: {len(ak_gpp.time)}")

# 检查空间坐标对齐
print("\n=== 空间坐标检查 ===")
print(f"LAI 使用的坐标: lat {ak_lai.latitude.values}, lon {ak_lai.longitude.values}")
print(f"GPP 使用的坐标: lat {ak_gpp.latitude.values}, lon {ak_gpp.longitude.values}")
```

```python
import xarray as xr
import matplotlib.pyplot as plt
import pandas as pd

# 1. 加载数据
store = 'gs://leap-persistent-ro/data-library/feedstocks/metaflux_feedstock/meatflux_monthly.zarr'
ds = xr.open_dataset(store, engine='zarr', chunks={})

store_lai = 'https://nyu1.osn.mghpcc.org/leap-pangeo-pipeline/MODIS_LAI/MODIS_LAI.zarr'
LAI = xr.open_dataset(store_lai, engine='zarr', chunks={})

# 2. 阿拉斯加格点
lat_ak = 65.0
lon_ak = -150.0

# 3. 提取GPP和RECO（先做这个！）
ak = ds.sel(latitude=lat_ak, longitude=lon_ak, method="nearest")
ak_GPP = ak['GPP']
ak_RECO = ak['RECO']

print("实际使用的格点坐标：", ak.latitude.values, ak.longitude.values)

# 4. 计算NEE（现在ak_RECO和ak_GPP已经定义了）
ak_NEE = ak_RECO - ak_GPP

# 5. 预处理LAI数据
def preprocess_lai_for_alaska(gpp_ds, lai_ds, lat_ak=65.0, lon_ak=-150.0):
    lai_monthly = lai_ds['lai'].resample(time='1M').mean()
    lai_regridded = lai_monthly.interp(
        lat=gpp_ds.latitude,
        lon=gpp_ds.longitude,
        method='linear'
    )
    common_time = slice('2002-01-01', '2020-12-31')
    lai_aligned = lai_regridded.sel(time=common_time)
    gpp_aligned = gpp_ds.sel(time=common_time)
    
    ak_lai = lai_aligned.sel(latitude=lat_ak, longitude=lon_ak, method="nearest")
    ak_gpp = gpp_aligned.sel(latitude=lat_ak, longitude=lon_ak, method="nearest")
    
    return ak_lai, ak_gpp

# 调用函数
ak_lai, ak_gpp_aligned = preprocess_lai_for_alaska(ds, LAI, lat_ak, lon_ak)

# 6. 对齐NEE时间
ak_NEE_aligned = ak_NEE.sel(time=slice('2002-01-01', '2020-12-31'))

# 7. 现在绘制NEE
plt.figure(figsize=(12, 6))
plt.plot(ak_NEE_aligned.time, ak_NEE_aligned, label='NEE (RECO - GPP)', 
         color='purple', linewidth=2)

plt.axhline(y=0, color='black', linestyle='--', alpha=0.5, label='Carbon Neutral')
plt.xlabel('Time')
plt.ylabel('NEE (gC m⁻² month⁻¹)')
plt.title('Monthly Net Ecosystem Exchange (NEE) at Alaska Grid Point')
plt.legend()
plt.grid(True)
plt.show()

# 8. 统计分析
print("=== NEE 统计分析 ===")
print(f"NEE 平均值: {ak_NEE_aligned.mean().values:.3f} gC m⁻² month⁻¹")
print(f"NEE 年净值: {ak_NEE_aligned.mean().values * 12:.3f} gC m⁻² year⁻¹")
print(f"碳汇比例: {(ak_NEE_aligned < 0).sum().values / len(ak_NEE_aligned) * 100:.1f}% 的时间为碳汇")

if ak_NEE_aligned.mean().values < 0:
    print("✅ 该生态系统总体为碳汇")
else:
    print("⚠️ 该生态系统总体为碳源")

# 9. 可视化GPP, RECO, NEE对比
plt.figure(figsize=(12, 8))

plt.subplot(2, 1, 1)
plt.plot(ak_GPP.time, ak_GPP, label='GPP', color='green', linewidth=2)
plt.plot(ak_RECO.time, ak_RECO, label='RECO', color='red', linewidth=2)
plt.ylabel('Carbon Flux (gC m⁻² month⁻¹)')
plt.legend()
plt.title('GPP and RECO at Alaska Site')
plt.grid(True)

plt.subplot(2, 1, 2)
plt.plot(ak_NEE.time, ak_NEE, label='NEE (RECO - GPP)', color='purple', linewidth=2)
plt.axhline(y=0, color='black', linestyle='--', alpha=0.5)
plt.ylabel('NEE (gC m⁻² month⁻¹)')
plt.xlabel('Time')
plt.legend()
plt.title('Net Ecosystem Exchange (NEE)')
plt.grid(True)

plt.tight_layout()
plt.show()
```
